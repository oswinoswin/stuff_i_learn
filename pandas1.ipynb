```python
import pandas as pd
import numpy as np
from random import random, randint

n = 99
k = 4
colorsmap = ['red', 'blue', 'green', 'yellow']
X0 = [randint(100, 999) for q in range(n)]
X1 = [randint(100, 999) for q in range(n)]
Y = [randint(0, k-1) for q in range(n)]
colors = [colorsmap[q] for q in Y]

df = pd.DataFrame({
	'X0': X0,
    'X1': X1,
	'Y': Y,
	'color': colors}, 
	index=range(n), 
	columns=['X0', 'X1', 'Y', 'color'])
```
